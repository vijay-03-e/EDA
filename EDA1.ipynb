# 1.Data Cleaning and Preparation:

```python

# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Impute missing values with the mean
df.fillna(df.mean(), inplace=True)

# Drop rows with missing values
df.dropna(inplace=True)

# Print the updated DataFrame
print(df)
```

```python
# Identify columns with mixed data types
mixed_cols = df.select_dtypes(include=['object', 'number']).columns

# Convert mixed-type columns to appropriate numerical types
for col in mixed_cols:
  df[col] = pd.to_numeric(df[col], errors='coerce')

# Check for remaining inconsistencies
df.info()
```

```python
# Detect and treat outliers if necessary.

import numpy as np

# Calculate the z-score for each numerical column
z_scores = (df - df.mean()) / df.std()
z_scores

# We take threshold value as 3
threshold=3

# Identify outlier indices
outlier_indices = np.abs(z_scores) > threshold

# Replace outliers with NaN
df[outlier_indices] = np.nan

# Fill missing values with the mean
df.fillna(df.mean(), inplace=True)
df
```

 # 3.Data Visualization



```python

# Create histograms or boxplots to visualize the distributions of various numerical variables of df

import matplotlib.pyplot as plt

# Create histograms for numerical variables
df.hist(figsize=(10, 8), bins=10)
plt.show()

# Create boxplots for numerical variables
df.boxplot(figsize=(10, 8))
plt.show()
```

```python
# Use bar charts to display the frequency of categories for categorical variables 
import matplotlib.pyplot as plt
categorical_cols = df.select_dtypes(include='object').columns
print(categorical_cols)
# Create bar charts for categorical variables
for i in df:
    # plot dimension are 15X7
    plt.figure(figsize=(6,4))
    df[i].value_counts().plot(kind='bar')
    plt.title(f'Bar-Graph of {i}')
    # Hide x-axis labels
    plt.xticks([])
    plt.show()
    
    
df.skew()
```
